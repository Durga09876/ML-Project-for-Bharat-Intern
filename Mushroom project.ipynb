```python
inp=df.drop("class",axis=1)
out=df["class"]
```

```python
X=pd.get_dummies(inp,drop_first=True)
y=out
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=True)
```

```python
# Adaboost Hp
```

```python
from sklearn.ensemble import AdaBoostClassifier
estimator=AdaBoostClassifier(random_state=True)

from sklearn.model_selection import GridSearchCV
param_grid={"n_estimators":list(range(1,30))}

modelhp=GridSearchCV(estimator,param_grid,cv=5,scoring="accuracy")
modelhp.fit(X_train,y_train)
modelhp.best_params_
```

```python
modelhp.best_estimator_.feature_importances_
```

```python
f_imp=pd.DataFrame(data=modelhp.best_estimator_.feature_importances_,index=X.columns,columns=["Feature_imp"])
imp_f=f_imp[f_imp["Feature_imp"]>0].index.to_list()
```

```python
imp_f
```

```python
X_imp=X[imp_f]
X_train,X_test,y_train,y_test=train_test_split(X_imp,y,train_size=0.8,random_state=True)

from sklearn.ensemble import AdaBoostClassifier
ad=AdaBoostClassifier(n_estimators=21,random_state=True)
ad.fit(X_train,y_train)

ypred_train=ad.predict(X_train)
ypred_test=ad.predict(X_test)

from sklearn.metrics import accuracy_score
print("Train accuracy:",accuracy_score(y_train,ypred_train))

from sklearn.model_selection import cross_val_score
print("cv:",(cross_val_score(ad,X_train,y_train,cv=5)).mean())
```

```python
print("Test accuracy:",accuracy_score(y_test,ypred_test))
```
